# the function which returns all permutation of a game

```python

# تعریف تابع کلی 

# این تابع یه بازی را میگیره همه ۵۷۶ حالت ممکنه را با دیتافریم میده
def Game_2_2(player1_payoff,player2_payoff):
#     این جایگشت هر حالت را میده- تابع  جایگشت حالت ها
    def States(main_list=('A','B','C','D')):
        all_per_game=[list(game) for game in permutations((char) for char in main_list) ]
        return all_per_game
    # EXAMPLE=('A','B','C','D','E')
    States()
    
    # اسم گذاری موقعیت ها
    def naming_s(the_name,name='name'):
        all_names=[]
        for char in range(1,len(the_name)+1):
            all_names.append('%s%s%i'%(name,'_',char))
#         all_two= [str(char)for char in all_names]
        return (all_names)
    naming_s(States(),'State')


    # ساخت دیتا فریم موقعیت ها
    naming_game=naming_s(States(),'State')
    df_states=pd.DataFrame(States(),index=naming_game)
    # df_states


    
    def Forms(player1_payoff,player2_payoff):
        all_forms= list(list(zip(r, p)) for (r, p) in (zip (repeat(player1_payoff), permutations(player2_payoff))))
        permuted=[]
#         این یه تیکه برای لیست کردن تاپل هایاولیه است که نمیدونم چجوری با راه بهتر ا ز اول لیستش کنم
        for char in all_forms:
            my_char= []
            for sub_char in char:
                my_char.append(list(sub_char))
            permuted.append(my_char)
        return (permuted)
#     Forms(player_1,player_2)
    
    naming_form=naming_s(Forms(player_1,player_2),'Form')
    df_forms=pd.DataFrame(list(Forms(player_1,player_2)),index=naming_form)
    # df_forms
    
    my_dict={}
    for index, row in df_forms.iterrows():
        my_dict[index]=[*row]
#         بجای یک و صفر و سه اینو گذاشتم 
    # my_dict
    
    whole_dict={}
    for key,value in my_dict.items():
        whole_dict[key]=States(my_dict[key])
    # whole_dict

    df_forms=pd.DataFrame(whole_dict,index=naming_s(States(),'State'))
    return(df_forms)




```

```python
# انکود میکنه . ورودی را میگیره . هر ورودی بود به کد ۰ تا ۴ تبدیل میکنه 
def standard_set(player1_payoff,player2_payoff):
    the_encode=LabelEncoder()
    feature_1=the_encode.fit_transform(player1_payoff)
    feature_2=the_encode.fit_transform(player2_payoff)
    natural_number_1=[number+1 for number in feature_1]
    natural_number_2=[number+1 for number in feature_2]
    both=(natural_number_1 ,natural_number_2)
    return list(both)


# این تابع هم سورت میکنه بازیکن ۱ و ۲ را تا بشه باهاش پیدا  کرد کدوم وضعیت ۶۴ گانه است
def set_standard_origin(player1,player2):
    sort_player1=player1.copy()
    sort_player2=player2.copy()
    sort_player1.sort()
    sort_player2.sort()
    both=(sort_player1 ,sort_player2)
    return list(both)

def set_code_standard(player1,player2):
    p1,p2=set_standard_origin(player_1,player_2)
    both= standard_set(p1,p2)
    return list (both)
# example:
# player1=[1,-1,-1,4]
# player2=[14,-8,-1,3]
# play1,play2=standard_set(player1,player2)
# set_standard_origin(play1,play2)
```

```python
# این تابع همه ۶۴  وضعیت ممکن . یعنی وضعیت مادر را میده که هر کدومشون ۵۷۶ بازی خواهند داشت
# یه دیتا فریم میده .  یا حتی میشه دیکشنری بده
# بعدا بهتره کلاسشو بنویسم که با متد صدا بزنم  دیکشنریشو میخوام یا.....
def Situation_all(number_game_and_strategy=2*2,rotation='pre'):
    equality_list_2_2=[
    [1,2,3,4],
    [1,1,2,3],
    [1,2,2,3],
    [1,2,3,3],
    [1,1,2,2],
    [1,1,1,2],
    [1,2,2,2],
    [1,1,1,1]
         ]
    product_feature=list(product(equality_list_2_2,equality_list_2_2))
#     product_feature
#     تابع اسم گذاری طبق ترتیب. کافیه بجای ایندکس اینو بدی و اون اسم را
    def naming_s(the_name,name='name'):
        all_names=[]
        for char in range(1,len(the_name)+1):
            all_names.append('%s%s%i'%(name,'_',char))
        #         all_two= [str(char)for char in all_names]
        return (all_names)
    df_db_feature=pd.DataFrame(product_feature,columns=['player_1','player_2'], index=(naming_s(product_feature,'Situation')))
    if rotation=='pre':
        return df_db_feature
    elif rotation=='transpose':
        return df_db_feature.transpose()
    elif rotation =='both':
        return df_db_feature,df_db_feature.transpose()

# Situation_all()
rotation=Situation_all(rotation='transpose')
origin=Situation_all()
rotation
```

```python

# تابعی که ورودی بگیره بگه برای کدوم سیچو ایشن هستش
def sitiatuion_determine(player_1,player_2,return_type='frame'):
    situation_dict=Situation_all(rotation='transpose').to_dict()
    situation_dict_extract={}
    for situation ,value in situation_dict.items():
        situation_dict_extract[situation]=[list(value.values())]
    # situation_dict_extract
    alldf=pd.DataFrame(situation_dict_extract,index=['players_payoffs_profile'])
#     alldf
    MY=set_code_standard(player_1,player_2)
#     MY
    if return_type =='stack':
        return alldf[alldf.isin([MY])].stack()
    elif return_type=='frame':
        Situation_location=alldf[alldf.isin([MY])]
        df=Situation_location.transpose()
        df2=df[df['players_payoffs_profile'].notnull()].transpose()
        return df2 



# # . نمونه ازمایشی
# player_1= [1, 1, -2, -2]
# player_2=[1, -2, -2, -2]
# a=sitiatuion_determine(player_1,player_2,return_type='frame')
# a
```
