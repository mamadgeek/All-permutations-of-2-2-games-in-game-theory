```python
import numpy as np
import pandas as pd
import itertools
from itertools import permutations, combinations,product,combinations_with_replacement,repeat
import scipy
```

```python
# این تابع به اندازه داده هایی که میدیم میاد لیستشون میکنه و جایگشتونو درست میکنه درون یک لیست از لیست ها میده 

def States(main_list=('A','B','C','D')):
    all_per_game=[ list(game) for game in permutations((char) for char in main_list) ]
    return all_per_game
# main=('A','B','C','D','E')
States()
```

```python
# این میاد به اندازه همه موارد, بازی ها را نام گذاری میکنه بهمون میده
# تابع نام گذاری روی یه سری چیز
# اسمشهم پیشفرض چیزی میدیم ولی میشه عوض کرد
def naming_s(the_name,name='name'):
    all_names=[]
    for char in range(1,len(the_name)+1):
        all_names.append('%s%s%i'%(name,'_',char))
    return (all_names)
naming_s(States(),'State')
```

```python
# این هر سطر یه بازی بشه که جایگاه هر کدام هر خودش نوشته میشه
naming_game=naming_s(States(),'State')
df_states=pd.DataFrame(States(),index=naming_game)
df_states
```

```python

# جایشگت همه فرم ها را میده
# اولی را ثابت میکنه دومی را میچرخونه 
# 24 حالت میده
player_1 ,player_2 =[1,2,3,4],[1,2,3,4]


list(list(zip(r, p)) for (r, p) in zip(repeat(player_1), permutations(player_2)))



# حالا هر کدوم از اینها یه فرم هستند 
# که باید هر فرم را به  موارد نسبت بدیم که در جدوله هستند
```

```python
# تابع فرم را درست میکنه
def Forms(player1_payoff=[1,2,3,4],player2_payoff=[1,2,3,4]):
    all_forms= list(list(zip(r, p)) for (r, p) in zip(repeat(player1_payoff), permutations(player2_payoff)))
    return (all_forms)
Forms()
```

```python
# اینو میاریم توی تابع اسم گذاری ترتیبی 
naming_s(Forms(),'Form')
```

```python
# اون تابع فرم را که درست کرده را به تابع نیم نسبت میده که اسم گذاری میکنه 
# و هر کدوم را یه تاپل میکنه میده
# تعداد خروجی های فایل فرمز را میگیره و اسم فرم میزاره روش
naming_form=naming_s(Forms(),'Form')
df_forms=pd.DataFrame(Forms(),index=naming_form)
df_forms
```
