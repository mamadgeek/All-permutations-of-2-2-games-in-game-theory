```python
import numpy as np
import pandas as pd
import itertools
from itertools import permutations
```

```python
# این تابع به اندازه داده هایی که میدیم میاد لیستشون میکنه و جایگشتونو درست میکنه درون یک لیست از لیست ها میده 
def Games(main_list=('A','B','C','D')):
    all_per_game=[ list(game) for game in permutations((char) for char in main_list) ]
    return all_per_game
# main=('A','B','C','D','E')
Games()
```

```python
# این میاد به اندازه همه موارد بازی ها را نام گذاری میکنه بهمون میده

def naming_games(the_game=Games()):
    all_games=[]
    for char in range(1,len(the_game)+1):
        name='GAME_'
        all_games.append('%s%i'%(name,char))
    return (all_games)
naming_games()
```

```python
# الان باید بیام دیتا فریم درست کنم ستون ها بشه  موارد بازی ها و هد بشه اسم بازی ها 

df=pd.DataFrame(Games())
df.T
```

```python
# این هر سطر یه بازی بشه که جایگاه هر کدام هر خودش نوشته میشه
df=pd.DataFrame(Games(),index=naming_games())
df
```

```python
# اینم هر ستون را درست میکنه . یعنی برعکس میکنیم . که هر ستون یه بازی بشه
df=pd.DataFrame(Games(),index=naming_games())
df.T
```
